##Preliminary Analysis on Categories and Finished/Unfinished Paths

###Categories

```python
categories = categ_df['Categories']
categories
```

```python
main_cat_list = []
sec_cat_list = []
thrd_cat_list = []
length = 0
for cat in categories:
  splitted = cat.split('.')
  if splitted[1] not in main_cat_list :
      main_cat_list.append(splitted[1])

  try:
    if splitted[2] not in sec_cat_list:
      sec_cat_list.append(splitted[2])
    if splitted[3] not in thrd_cat_list:
      thrd_cat_list.append(splitted[3])
  except:IndexError

  finally:
    pass

print(f'There are {len(main_cat_list)} unique main categories, {len(sec_cat_list)} unique second level categories, and {len(thrd_cat_list)} unique third level categories.')
```

####Discussion:
As we have 4604 articles with 15 main, 101 second level, 27 third level categories, we will be able to make more genera observations, if there is any, on how humans make connections between different categories.

###Finished & Unfinished Paths

```python
print(f"Although there are {len(paths_fin_df)} finished paths in the data, there are {len(paths_fin_df['Hashed IP Address'].unique())} players which means there \nare several finished paths belonging to the same player which might contain important information.")
```

```python
grouped_fin_df = paths_fin_df.groupby('Hashed IP Address')
duplica_fin_df = grouped_fin_df.filter(lambda x: len(x) > 1)
print(f" {len(duplica_fin_df['Hashed IP Address'].unique())} players have more than one finished paths in the data.")
```

```python
#example of a player that finished the game four times
paths_fin_df[paths_fin_df['Hashed IP Address'] == '6a3701d319fc3754']
```

```python
print(f"Following the same logic there are {len(paths_unfin_df)} unfinished paths and {len(paths_unfin_df['Hashed IP Address'].unique())} players in the data which means there \nare several unfinished paths belonging to the same player which might again contain important information.")
```

```python
grouped_unfin_df = paths_unfin_df.groupby('Hashed IP Address')
duplica_unfin_df = grouped_unfin_df.filter(lambda x: len(x) > 1)
print(f" {len(duplica_unfin_df['Hashed IP Address'].unique())} players have more than one unfinished paths in the data.")
```

```python
#example of a player that have 5 unfinished paths
paths_unfin_df[paths_unfin_df['Hashed IP Address'] == '6d136e371e42474f']
```

```python
fin_IP_set = set(paths_fin_df['Hashed IP Address'].unique())
unfin_IP_set = set(paths_unfin_df['Hashed IP Address'].unique())
players_fin_unfin = fin_IP_set.intersection(unfin_IP_set)
print(f"{len(players_fin_unfin)} players have both finished and unfinished paths which might contain important information about the success of the players.")
```
